```r
text<-readLines("11处特工皇妃.txt")
text<-text[nchar(text)!=0]
text[1:100]
```

```r
章节名<-(grep("第.{1,5}章|引子|引言|末章|楔子|后记|尾声",text,value=TRUE))
章节分段<-c(grep("第.{1,5}章|引子|引言|末章|楔子|后记|尾声",text),length(text)+1)
dir.create("C:/Users/Kismet/Documents/我的坚果云/小说/11处特工皇妃",recursive=TRUE)
分段章节<-list()
for (i in 1:(length(章节分段)-1)){
    data<-text[章节分段[i]:章节分段[i+1]-1]
    分段章节[[i]]<-data
    names(分段章节)[i]<-章节名[i]
    write(data,paste0("C:/Users/Kismet/Documents/我的坚果云/小说/11处特工皇妃/",i,".txt"))
}
```
